```python
# Function to identify fuzzy duplicates and mark them
def mark_fuzzy_duplicates(df, column_name, threshold=95):
    # Copy of the DataFrame to avoid modifying the original
    df_with_matches = df.copy()
    unique_elements = []
    df_with_matches['Match'] = None  # Initialize the Match column
    
    for i, value in enumerate(df[column_name]):
        # Attempt to find a match in the unique_elements list
        match = process.extractOne(value, unique_elements, score_cutoff=threshold)
        
        if not match:
            unique_elements.append(value)
        else:
            # If a match is found, store the matching value in the 'Match' column
            df_with_matches.at[i, 'Match'] = match[0]
    
    # Marking rows without any match as unique
    df_with_matches['Match'].fillna('Unique', inplace=True)
    
    return df_with_matches
```

```python
df_operating_name=df_wards[["Operating Name"]][~df_wards["Operating Name"].isna()].reset_index(drop=True)
df_operating_name
```

```python
# Applying the function
df_marked = mark_fuzzy_duplicates(df_operating_name, "Operating Name")

df_marked
```

```python
df_marked[df_marked["Match"]!="Unique"].loc[200:250]
```

```python
df_marked["business_operating_name"]=np.where(df_marked["Match"]=="Unique",df_marked["Operating Name"],df_marked["Match"])
df_marked
```

```python
df_marked.to_excel("formatted_op_names.xlsx")
```

```python
df_bus_fixed=df_bus.merge(df_marked[["Operating Name","business_operating_name"]],on="Operating Name",how="left")
df_bus_fixed
```

```python
df_bus_fixed["business_operating_name"]=df_bus_fixed["business_operating_name"].fillna(df_bus_fixed["Client Name"])
df_bus_fixed
```

```python
df_bus_fixed_subset.drop_duplicates(subset="business_operating_name",inplace=True)
```
